```python
import pymysql
import pandas as pd
import numpy as np
import pickle
import warnings
import locale
import re
locale.setlocale(locale.LC_ALL, '')
warnings.filterwarnings(action='ignore')
path = '' # 저장경로 지정이 필요합니다
```

```python
raw_data = pd.read_csv('./법정동코드 조회자료.csv', encoding='cp949')
raw_data
```

```python
def get_location_name(name_location):
    split_list = re.split(r'[ ]', name_location) # 띄어쓰기를 기준으로 문자열 나누기
    자치구 = split_list[1]
    법정동 = split_list[2]
    return 법정동, 자치구
```

```python
test = raw_data.loc[2, '법정동명']
법정동, 자치구 = get_location_name(test)
print(f'법정동 : {법정동}\n자치구 : {자치구}')
```

```python
len(raw_data.loc[1, '법정동명'])
```

```python
location_column = raw_data['법정동명']
법정동_list = []
자치구_list = []

for idx in range(len(location_column)):
    if len(location_column[idx]) < 11:
        continue
    else:     
        법정동, 자치구 = get_location_name(location_column[idx])
        법정동_list.append(법정동)
        자치구_list.append(자치구)

dictionary_location = dict(zip(법정동_list,자치구_list))
dictionary_location
```

# Pickle파일형태로 저장

```python
def save_pickle(name_pkl, obj):
    with open(name_pkl, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
# 예시
save_pickle('Seoul_dict.pickle', dictionary_location)
```

```python
## 우편번호 - 행정동 딕셔너리 형태로 만들기
def load_pickle(name_pkl):
    with open(name_pkl, 'rb') as f:
        obj = pickle.load(f)
    return obj
# 예시
data = load_pickle('경로/data.pickle')
## pickle 저장 코드 (dict1 객체 저장)
def save_pickle(name_pkl, obj):
    with open(name_pkl, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
# 예시
save_pickle('경로.pickle', data)
```
